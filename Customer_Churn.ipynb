```python
#Here we are using label Encoder for changing the Object type data to numerical data.
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df['Churn']=LE.fit_transform(df['Churn'])
df['Churn']
```

```python
#Now i have to find customer churn with respect to MonthlyCharges
```

```python
X=df[['MonthlyCharges']]
```

```python
Y=df[['Churn']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.35,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train,Y_train)
```

```python
model_predict=LR.predict(X_test)
model_predict
```

```python
#To Eveluate the model in logistic regression we use accuracy_score.
```

```python
from sklearn.metrics import accuracy_score
accuracy_model=accuracy_score(Y_test,model_predict)
accuracy_model
```

```python
sns.regplot(x=Y_test,y=model_predict)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Actual VS Predicted")
plt.grid(True)
plt.show()
```

```python
#Now we will use Decision Tree Classifier for Predicting the model
from sklearn.tree import DecisionTreeClassifier
```

```python
X=df[['tenure']]
Y=df[['Churn']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=12,random_state=12)
```

```python
model_predict=Decision_Model.predict(x_test)
model_predict
```

```python
acc=accuracy_score(y_test,model_predict)
acc
```

```python
sns.regplot(x=y_test,y=model_predict)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Actual vs Predicted")
plt.grid(True)
plt.show()
```
